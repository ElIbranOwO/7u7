```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-5,5,100)
```

```python
y=x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
y=1/x
```

```python
plt.plot(x,y)
plt.show()
```

```python
y=(x**2-3*x)/x
```

```python
plt.plot(x,y)
plt.show()
```

```python
y=np.tanh(x)/np.cosh(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
y=np.exp(-0.1*x)*np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```
